### Kiểm tra Account Number và Customer ID

```python
data['DUP_ACCT_NUM'] = data.duplicated(subset=['ACCT_NUM', 'REPORT_DATE'])
data[(data['DUP_ACCT_NUM'] == True)]
```

```python
data = data[data['DUP_ACCT_NUM'] == False]
```

### Kiểm tra Nhóm nợ

```python
data['NHOMNO'] = data['NHOMNO'].replace({'01': '1', '1': '1'})
data['NHOMNO'] = data['NHOMNO'].replace({'02': '2', '2': '2'})
data['NHOMNO'] = data['NHOMNO'].replace({'03': '3', '3': '3'})
data['NHOMNO'] = data['NHOMNO'].replace({'04': '4', '4': '4'})
data['NHOMNO'] = data['NHOMNO'].replace({'05': '5', '5': '5'})
```

```python
# Bin DPD
bins = [0, 1, 10, 30, 60, 90, np.inf]
labels = ['0', '01-10', '11-30', '31-60', '61-90', '91+']
data['DPD_BIN'] = pd.cut(data['DPD'], bins=bins, labels=labels, right=True, include_lowest=True)
```

```python
matrix = pd.crosstab(data['DPD_BIN'], data['NHOMNO'])
matrix.T
# => Có thể kiểm tra từng trường hợp dữ liệu đặc biệt để check lý do nhóm nợ nhảy không theo DPD
```

```python
# Đếm số nhóm nợ khác nhau cho mỗi CUST_ID và REPORT_DATE
group_counts = data.groupby(['CUST_ID', 'REPORT_DATE'])['NHOMNO'].nunique()
# Lọc các cặp (CUST_ID, REPORT_DATE) có nhiều hơn 1 nhóm nợ
multi_group = group_counts[group_counts > 1].reset_index()
# Join để lấy tất cả dòng có CUST_ID và REPORT_DATE nằm trong danh sách trên
dup_nhomno = data.merge(multi_group, on=['CUST_ID', 'REPORT_DATE'], how='inner')
dup_nhomno.sort_values(by=['CUST_ID', 'NHOMNO_x'])
# => Dự kiến giữ lại các cột có nhóm nợ cao nhưng DPD lại không khớp với nhóm nợ -> drop những khách hàng có DPD không khớp, cần kiểm tra thêm trên tập tổng
```

```python
# Loại bỏ các dòng trong result ra khỏi data
data = data.merge(dup_nhomno, how='left', indicator=True)
data = data[data['_merge'] == 'left_only'].drop(columns=['_merge'])
```

```python
# Lấy ngày gần cuối tháng nhất cho mỗi nhóm
latest_dates = data.groupby(['ACCT_NUM', 'YEAR', 'MONTH'])['REPORT_DATE'].max().reset_index()
data = data.merge(latest_dates, on=['ACCT_NUM', 'YEAR', 'MONTH', 'REPORT_DATE'], how='inner')
data = data.sort_values(['ACCT_NUM', 'YEAR', 'MONTH'])
# Aggregate MOB
data['MOB'] = (
    data.groupby(['ACCT_NUM'])['REPORT_DATE']
    .rank(method='dense')
    .astype(int) - 1  # bắt đầu từ 0
)
```
